**Discrete Uniform Distribution**- probability distribution where values are eaqual likely and they are finite values. 

Continuous Uniform Distribution- probability distribution where each value within a certain range is equally likely to occur and values outside the range never occur. There is an infinite number of points that can exist. 

Formula: image.png

```
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np
```

```
fig, ax = plt.subplots(1, 1)
x = np.linspace(stats.uniform.ppf(0.01),
                stats.uniform.ppf(0.99), 100)
ax.plot(x, stats.uniform.pdf(x),'r-', lw=5, alpha=0.6, label='uniform pdf')
r = stats.uniform.rvs(size=1000)
ax.hist(r, density=True, histtype='stepfilled', alpha=0.2)
ax.legend(loc='best', frameon=False)
plt.show()
```

```
uniform_data = stats.uniform.rvs(size=100000, loc = 10, scale=20)
#pd.DataFrame(uniform_data).plot()
pd.DataFrame(uniform_data).plot(kind='density')
```

Bernoulli Distribution- Descrete probability distribution of a random variable which takes a binary, boolean output. It has two possible outcomes 1 with probability p and 0 with probability (1-p)

Formula: bernoulli formula.PNG

```
p =  0.5 # parametro de forma 
bernoulli = stats.bernoulli(p)
x = np.arange(-1, 3)
fmp = bernoulli.pmf(x) # Funcin de Masa de Probabilidad
fmp
fig, ax = plt.subplots()
ax.plot(x, fmp, 'bo')
ax.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
ax.set_yticks([0., 0.2, 0.4, 0.6])
plt.title('Bernoulli Distribution')
plt.ylabel('probability')
plt.xlabel('values')
plt.show()
```

**Pareto**- 
*   Continuous porbability distribution 
*   Called the 80-20 rule
*   Used to model distribution of incomes. 80% of the wealth is owned by 20% of the people. 
*   Used to describe real-world problems (social, scientific, geographycal, actuarial)

Parameters: 


*   a= location parameter. Sets the lower limit of time
*   b= shape parameter. Affects the shape of the distribution
*   x= quantiles
